Goal here is to find the covid stats in the various neighborhoods around the Pittsburgh area. 

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [15,8]
#hourly air quality file 
df = pd.read_csv("covid.csv")
df.sort_values("infections", inplace=True)
index = df.iloc[:,3] < 3
plt.title("Covid infections by Neighborhood")
plt.barh(df["neighborhood_municipality"][-20:],df["infections"][-20:])
plt.show()
```
